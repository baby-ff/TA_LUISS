```python
import numpy as np
import matplotlib.pyplot as plt
```

# EX4

```python
#create a grid
w = 3
Y, X = np.mgrid[-w:w:100j, -w:w:100j]
#evaluate the function
U = 2*Y-X
V = Y**2-Y

plt.figure()
#streamplot
plt.streamplot(X, Y, U,V, density=[1,.8], linewidth=None, color=None, cmap=None)
#points and lines
plt.plot([0,2],[0,1],'ro')
plt.axhline(y=0,lw=1,color='k')
plt.axvline(x=0,lw=1,color='k')
plt.axhline(y=1,lw=1,ls='-.',color='C0')
plt.plot([1.5,2.5],[.5,1.5],'-.',color='C0')
plt.show()
```

# EX3 - first integral

```python
#create a grid
w = 2.5
Y, X = np.mgrid[-w:w:100j, -w:w:100j]
#evaluate the function
U = Y
V = X-X**3

plt.figure()
#streamplot
plt.streamplot(X, Y, U,V, density=[1.2,.8], linewidth=None, color=None, cmap=None)
#points and lines
plt.plot([0,1,-1],[0,0,0],'ro')
plt.axhline(y=0,lw=1,color='k')
plt.axvline(x=0,lw=1,color='k')
plt.show()
```

```python
#compare with level curves of the first integral

#create a grid
w = 2.5
Y, X = np.mgrid[-w:w:100j, -w:w:100j]
def n(x):
    return x**2
#evaluate the function
H = .5*Y*Y-.5*X*X+.25*X**4
plt.figure()
plt.contour(X,Y,H,levels=-.25+n(np.linspace(0,4,30)))
#points and lines
plt.plot([0,1,-1],[0,0,0],'ro')
plt.axhline(y=0,lw=1,color='k')
plt.axvline(x=0,lw=1,color='k')
plt.title('level curves of $H(x,y)$')
plt.show()
```

# SIR with birth-death

```python
#play with SIR parameters
beta=1
mu=.5
gamma=.5
R0=beta/(mu+gamma)
#create a grid
Y, X = np.mgrid[0:1:100j, 0:1:100j]
#evaluate the function
U = -beta*X*Y-mu*(X-1)
V = beta*X*Y - (mu+gamma)*Y

plt.figure(figsize=(7,7))
plt.xlim(0,1)
plt.ylim(0,1)
plt.xlabel('S/N')
plt.ylabel('I/N')
#streamplot
plt.streamplot(X, Y, U,V, density=[1.3,1.3], linewidth=None, color=None, cmap=None)
plt.fill([1,1,0],[1,0,1],'grey',zorder=3)
#eq points
plt.plot(1,0,'ro',clip_on=False)
if(R0>1):
    plt.plot(1/R0,mu*(R0-1)/beta,'ro',clip_on=False)
plt.title(r'$R_0$=%.2f'%R0)
plt.show()
```
